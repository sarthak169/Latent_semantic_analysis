```python
l = []
i = []
for file in glob.glob('./*.txt'):
    with open('./final_Data/'+file.replace('txt','pdf'), 'rb') as pdfFileObj:
        pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
        info = pdfReader.getDocumentInfo()
        i.append(info)
    with open(file,encoding = 'UTF-8') as f:
        a = f.read()
        l.append(a)
d = {'metadata':i,'text':l}
df = pd.DataFrame(d)
```
